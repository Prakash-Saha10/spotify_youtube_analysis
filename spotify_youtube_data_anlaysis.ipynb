# **Top 10 Artists - with the Highest Views on YouTube?**

```python
artist_grouped= data.groupby('Artist')['Views'].sum()
artist_grouped
```

```python
artist_sorted=artist_grouped.sort_values(ascending=False)
artist_sorted
```

**10 Tracks - with the Highest Streams on Spotify?**

```python
x=data[['Stream','Track']]
x
```

```python
most_popular_track=x.sort_values('Stream',ascending=False)
most_popular_track
```

```python
#5 Tracks - with the Lowest Streams on Spotify?

least_popular_track=x.sort_values('Stream',ascending=True).head(5)
least_popular_track

```

**What are the most common Album Types on Spotify? How many tracks belong to each album typet**

```python
album_list=data['Album'].sort_values(ascending=False).head(10)
album_list
```

```python
album_type=data['Album_type'].value_counts()
album_type
```

```python
plt_pie=album_type.plot.pie(autopct='%1.1f%%',figsize=(5,5),explode=(0.05,0.05,0.05),pctdistance=0.85,shadow=True)
plt_pie

#Explode – To cut the slices out.
#Autopct – To show the % on the chart using string format.
#pctdistance – Distance of % from center
```

** Top 5 YouTube Channels - based on the Views?**

```python
c_views = data.groupby('Channel')['Views'].sum().sort_values(ascending=False).head()

c_views
```

```python
c_views = c_views.reset_index()

c_views.head(10)
```

```python
# sns.set_style("whitegrid")

sns.barplot( x = "Views", y = "Channel", data = c_views, color='black')
plt.title('Top 5 Channels by Views')
plt.xlabel('Views')
plt.ylabel('Channel')
plt.show()
```
